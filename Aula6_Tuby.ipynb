```python
import matplotlib.pyplot as plt
import numpy as np

def retas_paralelas():
    x1=np.arange(10, 90, 20)
    x2=np.arange(20, 100, 20)

    y1=np.array(x1)/100

    plt.plot(x1,y1)
    plt.plot(x1,y1, 'go')
    plt.plot(x2,y1)
    plt.plot(x2,y1, 'ro')

    plt.show()
    
def retas_concorrentes():
    x1=np.arange(10, 90, 20)
    x2=np.arange(80, 0, -20)

    y1=np.array(x1)/100

    plt.plot(x1,y1)
    plt.plot(x1,y1, 'go')
    plt.plot(x2,y1)
    plt.plot(x2,y1, 'ro')

    plt.show()
    
def retas():
    x1=np.linspace(10,40,4)
    y1=np.array(x1)/100
    
    x2=x1
    y2=y1
    
    plt.plot(x1,y1)
    plt.plot(x1,y1, 'go')
    plt.plot(x2,y2)
    plt.plot(x1,y1, 'ro')
    
    plt.show()
    
retas_paralelas()
retas_concorrentes()
retas()
```
